```python
spam_or_not=df.Message_body
spam_or_not
```

```python
#removing spcl chr
spam_or_not=spam_or_not.str.replace("[^a-zA-Z0-9]+"," ")
spam_or_not
```

```python
#stemming
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer
sb=SnowballStemmer('english')
tk=TweetTokenizer()
```

```python
spam_or_not=spam_or_not.apply(lambda x:[sb.stem(i.lower())for i in tk.tokenize(x)]).apply(lambda i:" ".join(i))
spam_or_not
```

```python
#remove stop words
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
sw=stopwords.words('english')
sw
```

```python
spam_or_not=spam_or_not.apply(lambda x:[i for i in tk.tokenize(x) if i not in sw]).apply(lambda i:" ".join(i))
spam_or_not
```

```python
#vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(spam_or_not)
print(X)
```

```python
vectorizer.get_feature_names_out()
```

```python
X.shape
```

```python
y=df['Label']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=14)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
knn=KNeighborsClassifier(n_neighbors=11)
sv=SVC()
rf=RandomForestClassifier(random_state=14)
ad=AdaBoostClassifier(random_state=14)
```

```python
from sklearn.metrics import classification_report
model=[knn,sv,rf,ad]
for i in model:
  print("***********",i,"**************")
  i.fit(X_train,y_train)
  ypred=i.predict(X_test)
  print(classification_report(y_test,ypred))
```
